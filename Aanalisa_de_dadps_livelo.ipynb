```python
#Construindo a funcao para blotar gráfico de barras

def graf_barra(df, coluna,tipo = 'porcentagem', figsize = 'normal'):

  df_print = pd.DataFrame(df.groupby(coluna).count()['Contagem'])

  if tipo == 'porcentagem':
    df_print['porcentagem'] = (df_print['Contagem']/df_print['Contagem'].sum())*100
  else:
    df_print['porcentagem_do_total'] = (df_print['Contagem']/len(df))*100
  labels = df_print.index
  contagem = df_print['Contagem']
  porcentagem = df_print[tipo]

  if figsize == 'normal':
    fig, ax = plt.subplots(1,1,figsize= (8,len(df_print.index)))
  else:
    fig, ax = plt.subplots(1,1,figsize= figsize)
  ax2 = ax.twiny()
  ax2.set_ylabel('Porcentagem', color='b')
  ax.barh(labels, contagem, label= porcentagem)
  ax2.barh(labels, porcentagem, label=contagem)
  bar_leg = ax.containers[0]
  ax.bar_label(bar_leg, labels=[f'{x:,.0f}' for x in bar_leg.datavalues])
  ax.set_xlabel('Número de Observacoes')
  ax2.set_xlabel('Porcentagem (%)')
  ax.set_title(coluna)
  plt.tight_layout()
  return plt.show()



```

```python
# Plotando as frequências

for coluna in df.columns[6:15]:
  #print(df_print)
  graf_barra(df,coluna)
```

```python
#cria a funcao Q_absoluto:

def Q_absoluto(df,nome_coluna, lista_de_colunas,lista_de_alternativas, Total = True):
  
  Q = []
  Q = pd.DataFrame(Q)
  Q[nome_coluna] = lista_de_alternativas
  for col in lista_de_colunas:
    row_list = []
    for programa in Q[nome_coluna]:
      novo_val = df.loc[df[nome_coluna].str.contains(programa)==True][col].sum()
      row_list.append(novo_val)
    Q[col] = row_list
  Q = Q.set_index(nome_coluna)
  if Total
    Q.loc["Total"] = Q.sum()
  return Q


#def Q_relativo(Q): Fayer uma funcao paracalcular a frequencia relativadepois


```

```python
#  Funcao Mapa de Medias

def mapa_medias(df):
  #gerando lista de medias
  media_ponderada_colunas =  pd.DataFrame(df[df.columns[30:175]].mean())
  # Construindo a matriz
  matriz_de_medias = pd.DataFrame(columns = Q_programas.keys(), index =score_dict.keys() )
  for linha in matriz_de_medias.index:
    for coluna in matriz_de_medias.columns:
      matriz_de_medias.at[linha,coluna] = media_ponderada_colunas.loc[(media_ponderada_colunas.index.str.contains(linha)==True) & (media_ponderada_colunas.index.str.contains(coluna)==True), 0][0]
  #matriz_de_medias.loc["Total"] = matriz_de_medias.sum()
  return matriz_de_medias

```

```python
#  Funcao Mapa de Valor

def mapa_de_valor(df):
  #gerando lista de medias
  media_ponderada_colunas =  pd.DataFrame(df[df.columns[30:175]].mean())
  # Construindo a matriz
  mapa_de_valor = pd.DataFrame(columns = Q_programas.keys(), index =score_dict.keys() )
  for linha in mapa_de_valor.index:
    for coluna in mapa_de_valor.columns:
      mapa_de_valor.at[linha,coluna] = media_ponderada_colunas.loc[(media_ponderada_colunas.index.str.contains(linha)==True) & (media_ponderada_colunas.index.str.contains(coluna)==True), 0][0]*score_dict[linha]
  mapa_de_valor.loc["Total"] = mapa_de_valor.sum()
  return mapa_de_valor
```
