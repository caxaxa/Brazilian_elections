```python
#Construindo a funcao para blotar gráfico de barras

def graf_barra(df_print,coluna,tipo = 'porcentagem'):
  fig, ax = plt.subplots(1,1,figsize=(8,len(df_print.index)))
  
  labels = df_print.index
  contagem = df_print['Contagem']
  porcentagem = df_print[tipo]

  ax2 = ax.twiny()
  ax2.set_ylabel('Porcentagem', color='b')
  ax.barh(labels, contagem, label= porcentagem)
  ax2.barh(labels, porcentagem, label=contagem)
  bar_leg = ax.containers[0]
  ax.bar_label(bar_leg, labels=[f'{x:,.0f}' for x in bar_leg.datavalues])
  ax.set_xlabel('Número de Observacoes')
  ax2.set_xlabel('Porcentagem (%)')
  ax.set_title(coluna)
  plt.tight_layout()
  return plt.show()



```

```python
# Plotando as frequências

df['Contagem'] = 1

for coluna in df.columns[6:15]:
  df_print = pd.DataFrame(df.groupby(coluna).count()['Contagem'])
  df_print['porcentagem'] = (df_print['Contagem']/df_print['Contagem'].sum())*100
  df_print['porcentagem_do_total'] = (df_print['Contagem']/len(df))*100
  #print(df_print)
  graf_barra(df_print,coluna)
```

```python
# Plotando as frequências ()

df['Contagem'] = 1

for coluna in df.columns[19:24]:
  df_print = pd.DataFrame(df.groupby(coluna).count()['Contagem'])
  df_print['porcentagem'] = (df_print['Contagem']/df_print['Contagem'].sum())*100
  df_print['porcentagem_do_total'] = (df_print['Contagem']/len(df))*100
  #print(df_print)
  graf_barra(df_print,coluna,'porcentagem_do_total' )
```
